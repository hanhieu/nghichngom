```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin
import re
from pdfminer.pdfparser import PDFParser
from pdfminer.pdfdocument import PDFDocument
import PyPDF2
```

# 1. Semper Augustus

```python
df = pd.DataFrame()
pdf = []
titles = []
url = 'https://www.semperaugustus.com/clientletter'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
#find .pdf in soup
for link in soup.find_all('a'):
    if link.get('href').endswith('.pdf'):
        pdfs = ('https:'+link.get('href'))
        title = (link.text)
        pdf.append(pdfs)
        titles.append(title)
    df = pd.DataFrame({'Title': titles, 'URL': pdf})
#remove special characters from title, : and ; and / to ''
df['Title'] = df['Title'].str.replace(':', '')
df['Title'] = df['Title'].str.replace(';', '')
df['Title'] = df['Title'].str.replace('/', '')
df
```

```python
#get oath to E drive
save_folder = os.chdir('E:/')
#save the pdf link in df to E:/fun read/Semper Augustus/superinvestors
for i in range(len(df)):
    response = requests.get(df['URL'][i])
    with open(f'E:/fun read/superinvestors/Semper Augustus/{df["Title"][i]}.pdf', 'wb') as f:
        f.write(response.content)
        f.close()
```

# 2. HOWARDS MARK

```python
url = "https://www.oaktreecapital.com/insights/memos"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
list_link = []
titles = []
url_pattern = r"https?://[^()'\"\s]+"
for link in soup.find_all('a', href=True):
    if '.pdf' in link.get('href'):
        title = (link.text)
        #get all the link of link.get('href) using regex
        urls = re.findall(url_pattern, link.get('href'))
        urls = [url for url in urls if '.pdf' in url]
        list_link.append(urls)
        titles.append(title)
    df = pd.DataFrame({'Title': titles, 'URL': list_link}).iloc[:-1]
df
       
        
```

```python
#get oath to E:/fun read/superinvestors/Howard Marks
save_folder = os.chdir('E:/fun read/superinvestors/Howard Marks')
files = os.listdir('E:/fun read/superinvestors/Howard Marks')
#compare the files in the folder and the files in the df, return the files that are not in the folder
df = df[~df['Title'].isin(files)]
```

```python
#save the pdf link in df to E:/fun read/superinvestors/Howard Marks
for i in range(len(df)):
    response = requests.get(df['URL'][i][0])
    with open(f'E:/fun read/superinvestors/Howard Marks/{df["Title"][i]}.pdf', 'wb') as f:
        f.write(response.content)
        f.close()
```

# 3. HAYDEN CAPITAL

```python
url = "https://www.haydencapital.com/wp-content/uploads/Hayden-Capital-Quarterly-Letter-"
list_link = []
list_name = []
for i in range(2017,2025):
    for j in range(1,5):
        link = url+str(i)+"-Q"+str(j)+".pdf"
        list_link.append(link)      
        file_name = os.path.basename(link)
        list_name.append(file_name)
df = pd.DataFrame({'URL': list_link, 'Title': list_name})
```

```python
def download_pdf(url, save_path):
    headers = {"Accept": "application/pdf",
              "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"}  # Specify that you accept PDF format

    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        with open(save_path, 'wb') as pdf_file:
            pdf_file.write(response.content)
        print(f"PDF downloaded successfully to {save_path}")
    elif response.status_code == 406:
        print("The server cannot provide a response in the format you accept (PDF).")
    else:
        print(f"Failed to download PDF. Status code: {response.status_code} at url: {url}")
```

```python
#get oath to E:/fun read/superinvestors/Hayden Capital
save_folder = os.chdir('E:/fun read/superinvestors/Hayden Capital/')
files = os.listdir(save_folder)
#compare the files in the folder and the files in the df, return the files that are not in the folder
df = df[~df['Title'].isin(files)]
```

```python
for i in range(len(df)):
    save_folder = f'E:/fun read/superinvestors/Hayden Capital/{df["Title"][i]}'
    download_pdf(df.loc[i,'URL'], save_folder)
```
